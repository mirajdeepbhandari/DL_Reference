```python
# Open the text file in read mode
with open('letters.txt', 'r') as file:
    # Read the contents of the file into a variable
    file_contents = file.read()

# Print the contents of the file
print(file_contents)
```

```python
file_contents = file_contents.replace("[Supplier's Name],", "")
file_contents = file_contents.replace("Dear", "")
file_contents = file_contents.replace("Hello", "")
file_contents
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([file_contents])
```

```python
tokenizer.word_index
```

```python
sentences_fil=[]
for sentences in file_contents.split('\n'):
  sentences_fil.append(sentences)
```

```python
final_sentences=[i for i in sentences_fil if len(i)> 1]
final_sentences
```

```python
final_sent="".join(final_sentences)
```

```python
final_preprocessed_text=final_sent.replace(",[Your Name]","")
final_preprocessed_text
```

```python
import re

for i in re.split(r'[.?]', final_preprocessed_text): # . ? ko adar ma split garxa

  final_preprocessed_texts=i.strip()
  print(final_preprocessed_texts)
```

```python
final_preprocessed_texts
```

```python
for eachline in re.split(r'[.?]', final_preprocessed_text):
  preprocessed_mails=tokenizer.texts_to_sequences([eachline])[0]
  print( preprocessed_mails)
```

```python
input_sequences=[]
for final_sentence in re.split(r'[.?]', final_preprocessed_text):

  tokenized_sentence=tokenizer.texts_to_sequences([final_sentence])[0]

  for i in range(1,len(tokenized_sentence)):

      input_sequences.append(tokenized_sentence[:i+1])

input_sequences
```

```python
All_length=[len(x) for x in input_sequences ]
print(All_length[:5])
max_len=max(All_length)
print("The maximum size of input sentence vector is : ",max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_inputs_sequences=pad_sequences(input_sequences, maxlen=max_len, padding='pre')
padded_inputs_sequences
```

```python
X=padded_inputs_sequences[:,:-1]
X
```

```python
y=padded_inputs_sequences[:,-1]
y
```

```python
print('X shape = ',X.shape)
print('y Shape = ',y.shape)
```

```python

from tensorflow.keras.utils import to_categorical

no_of_classes = len(tokenizer.word_index)
print("The number of classes are:", no_of_classes)

# Assuming y is your target variable
y = to_categorical(y, num_classes=no_of_classes + 1)  # to_categorical ley 0 bata ganxa so +1

print('New y shape:', y.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout


model = Sequential()
model.add(Embedding(410, 250, input_length=21))# total vocab counts, how many vectors each word be represented,
# All_length= 22 ,0 bata counting hunxa so -1 #The maximum size of input sentence vector is :  22 euta word 22 vector ley represent grxa
model.add(LSTM(500))
model.add(Dropout(0.5))
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(411, activation='softmax')) # jati ota class xa teti ota hunu paryo or no of vocabs

# Compiling the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


# Train the model with early stopping and validation split
history = model.fit(X, y, epochs=100, validation_split=0.2, verbose=1)
```

```python
import time
import numpy as np

text = input("Enter the text: ")

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])
  # padding
  padded_token_text = pad_sequences(token_text, maxlen=21, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
